# A8

## Iterating over characters in a file:

```python
try:
    stream = open('animals.txt')
    character = (stream.read(1))
    while character != '':
        print(character,end='_')
        character = stream.read(1)
    stream.close()
except Exception as e:
    print('An error has occured',e)
```

# A9
## Introducing a counter variable to iterating:


```python
try:
    stream = open('animals.txt')
    counter = 0
    character = (stream.read(1))
    while character != '':
        counter += 1
        print(character,end='_')
        character = stream.read(1)
    stream.close()
    print("\nChars in the file:",counter)
except Exception as e:
    print('An error has occured',e)
```

# A10 

## Rerading a file line by line:

```python
try:
    stream = open('animals.txt')
    counter = 0
    line = (stream.readline())
    while line != '':
        counter += 1
        print(line,end='\nLINE!\n')
        line = stream.readline()
    else:
        print("\nLINE!")
    stream.close()
    print("\nLines in the file:",counter)
except Exception as e:
    print('An error has occured',e)
```
