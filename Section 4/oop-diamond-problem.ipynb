### Two methods override each other

```python
class Vehicle:
    def show_power_type(self):
        print('Uses multiple power sources')

class ElectricVehicle(Vehicle):
    def show_power_type(self):
        print('Uses electricity')

class PetrolVehicle(Vehicle):
    def show_power_type(self):
        print('Uses petrol')

class HybridVehicle(ElectricVehicle,PetrolVehicle):
        pass
        
```

```python
i30 = HybridVehicle()
```

```python
i30.show_power_type()  # left side method resolution
```
