# write()

```python
try: 
    stream = open('list_of_cars.txt','w')
    stream.write('mazda\n')
    stream.write('toyota\n')
    stream.write('dodge\n')
    stream.close()
except Exception as e:
    print("An error of: ", e)
    
    
```

```python
try:
    stream = open('list_of_cars.txt')
    print(stream.read())
    stream.close()
except Exception as e:
    print("an error occured:",e)
```
