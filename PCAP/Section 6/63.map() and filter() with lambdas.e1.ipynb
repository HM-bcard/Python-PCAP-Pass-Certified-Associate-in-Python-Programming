# A1

## iterators

```python
lambda_1 = lambda i: i * 2
list1 = [i for i in range(1,6)]
```

# A2

## using a lambda fn in a map syntax:

```python
map(lambda_1,list1)
```

# A3 

## why isn't it displayed?

### Because map() is just an iterator that has to be ascribbed to a var

```python
map_elem = map(lambda_1,list1)
print(next(map_elem))
```

# A4

## Iterating:

```python
print(next(map_elem))
print(next(map_elem))
```

```python
for i in map_elem: print(i,end='!') # one liner!
```

# A5

## Putting all of the invocations into a single line:

```python
print(list(map(lambda i: i * 2,[i for i in range(1,6)])))
```

# A6 

## filter using lambda:

```python
print(list(filter(lambda i: i  == 2,[i for i in range(1,6)])))
```
