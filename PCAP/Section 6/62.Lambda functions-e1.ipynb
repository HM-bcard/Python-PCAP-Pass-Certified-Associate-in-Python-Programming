# A3

## Recursive function call:

```python
def apply(inp,fn):
    for el in inp:
        print(fn(el))
```

```python
fn1 = lambda x: x* x
apply(range(6),fn1)
```

```python
try:
    fn2 = lambda x: 1/ x
    apply(range(6),fn2)
except:
    apply(range(6),fn2)
```
