# A1

## What is a closure?

### A closure is a technique for implementing lexically scoped name binding in a language with first-class functions.

# A2 

## Explanation:

### A closure is a function defined another function that remembers the values of the outer function

# A3

## Example:

```python
def alarm(alarm_kind):

    def sound_alarm():
        print(alarm_kind)
    return sound_alarm
```

```python
alarm('fire')
```

```python
fire = alarm('fire')
```

```python
fire()
```

# A6

## Such variables are called free variables

```python
def multiply_c(number):

    def multiply(number_2):
        return number * number_2
    return multiply

multiply_5 = multiply_c(5)# moment it's initialized and stored - only one argument
multiply_12 = multiply_c(12)

print(multiply_5(5)) # second argument passed
print(multiply_5())
print(multiply_12(5))
```
