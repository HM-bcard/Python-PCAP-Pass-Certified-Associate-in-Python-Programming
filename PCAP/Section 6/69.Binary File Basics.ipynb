## What is the difference between text and binary files?

Different binary encoding - in a text file each byte represents a single character on an encoding table/plane ,
in other files different bits represents colours of pixels, or sounds etc

# A5 Saving Bytearrays into a file:

```python
file_1 = bytearray(100)
file_1[0] = 255
file_1[1] = 7
```

```python
try:
    stream = open('file_1.bin', 'wb') # write binary
    stream.write(file_1)
    stream.close()
except Exception as e:
    print('An error occured',e)
```

# A6 Opening a binary file:

```python
try:
    stream = open('file_1.bin','r')
    print(stream)
    stream.close()
except Exception as e:
    print('An error occured',e)
```

```python
try:
    stream = open('file_1.bin','rb')
    byte_array1 = stream.read()
    stream.close()
except Exception as e:
    print('An error occured',e)

print(byte_array1)
```

# A7 Hexadecimal type casting:

```python
for byte in byte_array1:
    print(hex(byte),end = '')
```

# A8 Decimal type casting:

```python
for byte in byte_array1:
    print(int(byte),end = '')
```

# A9 Pre-setting the maximal number of characters to be read:

```python
try:
    stream = open('file_1.bin','rb')
    byte_array1 = stream.read(10)
    stream.close()
except Exception as e:
    print('An error occured',e)

print(byte_array1)
```
