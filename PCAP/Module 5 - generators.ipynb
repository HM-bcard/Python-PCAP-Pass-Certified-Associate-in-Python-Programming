```python
def get_number():
    for i in range(1,4):
        yield i # keyword'yield' creates a generator
```

```python
print(get_number())
```

```python
generator = get_number()
```

```python
print(generator)
```

```python
print(next(generator))
```

```python
print(next(generator))
```

```python
for i in range (1 , 6):
    print(next(generator))
    
```
