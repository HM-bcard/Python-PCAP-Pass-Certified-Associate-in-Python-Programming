```python
input_numbers = [1,2,3,4,5,6,]

sum = 0.0
for x in input_numbers:
    sum += x
average = sum / len(input_numbers)
print(average)
```

```python
#input_numbers = [1,2,3,4,5,6,]

def get_average(input_numbers):
    sum = 0.0
    for x in input_numbers:
        sum += x
    average = sum / len(input_numbers)
    print(average)
    
```
