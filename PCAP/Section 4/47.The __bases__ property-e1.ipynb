## The \_\_bases\_\_ property - used to display all direct superclasses of a class

## Syntax:

```python
class Stationery():
    def write(self):
        print("blablabla")

    def what_am_i(self):
        print("i am a ",Stationery.__name__)

class Cylinder():
    def shape(self):
        print("i am cylindrical")

    def what_am_i(self):
        print("i am a ",Cylinder.__name__)

class Pen(Cylinder,Stationery): # object->right->left
    def what_am_i(self):
        print("i am a ",Pen.__name__)
```

```python
parker = Pen()
parker.what_am_i() 
```

```python
print(Pen.__bases__)
```
