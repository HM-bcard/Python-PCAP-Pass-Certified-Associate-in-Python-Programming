# A1

## Exemplary class

```python
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
```

# A2

## creating new properties per an object

```python
dog = Dog('Mike',21)
dog.eyes = 'blue'
```

```python
print(dog.eyes)
```

# A3
## deleting an existing property of an object

```python
del dog.age
```

```python
print(dog.age)
```

# A5

## accessing a private property

```python
class Dog:
    def __init__(self,name,age):
        self.__name = name
        self.age = age
```

```python
dog2 = Dog('Mike',22)
```

```python
dog2.__dict__
```

### this is called a mangled value

# A6

## accessing a mangled value:

```python
print(dog2._Dog__name)
```
