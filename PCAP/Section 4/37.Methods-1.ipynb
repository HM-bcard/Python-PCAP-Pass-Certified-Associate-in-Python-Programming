```python
class Engineer:
    def __init__(self,firstn = 'John',lastn = 'Smith'):
        
        self.firstn = firstn
        self.lastn = lastn
        self.__format()

    def __format(self):
        self.firstn = self.firstn.title()
        self.lastn = self.lastn.title()

    def present(self):
        print("Hello my name is ",self.firstn," ",self.lastn)

    def compare(self,name):
        if self.firstn == name:
            print("We both are ",self.firstn.upper(),"s" )
        else:
            print("Well i'm ",self.firstn)

    def fulln(self):
        return self.firstn+" "+self.lastn


engineer1 = Engineer("MiYaMoTo", "Musashi")
engineer1.present()
engineer1.compare("Takahoshii")
print(engineer1.__dict__)
print(Engineer.__dict__)
    
```
