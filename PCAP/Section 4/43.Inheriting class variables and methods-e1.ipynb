# A2 

## With the super() keyword

```python
class Animal:
    sound = "A generic animal sound"
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
    def __init__(self,speed,paws):
         super().__init__(speed)
         self.paws = paws
         print(super().sound)
    
    def run(self):
        self.speed += 2
        
class Dog(LandAnimal):
    def run_v_fast(self):
        print("woof woof woof ")
        super().run()
        super().run()
        super().run()
        super().run()
```

```python
mike = Dog(12,3)
print(mike.__dict__)
mike.run_v_fast()
print(mike.__dict__)
```
