# A2

## changing a superclass

```python
class Animal:
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
     pass  
class Dog(LandAnimal):
    pass
```

```python
charlie = Dog() # brackets required
print(charlie.__dict__)
```

### Explanation: if there is no constructior in a subclass but there is one in the superclass - python authomatically requires the constructor arguments

```python
charlie = Dog(10) # brackets required
print(charlie.__dict__)
```

```python
class Animal:
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
     def __init__(self,paws):
         self.paws = paws
class Dog(LandAnimal):
    pass
```

```python
charlie = Dog(10) # Python looks for the most proximate constructor - therefore uses the constructor of the direct superclass
print(charlie.__dict__)
```

## invoking the superclass conbstructor directly when instantiating

```python
class Animal:
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
     def __init__(self,speed,paws):
         Animal.__init__(self,speed)
         self.paws = paws
class Dog(LandAnimal):
    pass
```

```python
charlie = Dog(10,4) # Python looks for the most proximate constructor - therefore uses the constructor of the direct superclass
print(charlie.__dict__)
```

# A6

## super() notation

```python
class Animal:
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
     def __init__(self,speed,paws):
         super().__init__(speed)
         self.paws = paws
class Dog(LandAnimal):
    pass
```

```python
charlie = Dog(10,4) # Python looks for the most proximate constructor - therefore uses the constructor of the direct superclass
print(charlie.__dict__)
```
