# OOP example:

```python
class Rectangle:
    def __init__(self,width,height): # method
        self.width = width
        self.height = height
    def get_area(self): # method
        return self.width * self.height

a = int(input('What is the width?'))
b = int(input('What is the height?'))

rectangle = Rectangle(a,b)
print("Area:",rectangle.get_area())

age = int(input('What is your age?'))

rectangle2 = Rectangle(age,b)
print("Area:",rectangle2.get_area())
```
