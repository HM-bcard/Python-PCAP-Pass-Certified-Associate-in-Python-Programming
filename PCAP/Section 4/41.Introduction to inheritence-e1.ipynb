## Creating a super classes and sub classes

```python
class Animal:
    pass
```

```python
class LandAnimal(Animal):
     pass   
```

```python
class Dog(LandAnimal):
    pass
```

```python
issubclass(Dog,LandAnimal) # used to check if the class is a subclass of the second passed class
```

```python
issubclass(Dog,Animal)
```
