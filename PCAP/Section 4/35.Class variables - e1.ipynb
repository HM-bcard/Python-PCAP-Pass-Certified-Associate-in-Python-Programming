# A1

## Class usage

```python
class Dog:
    counter = 0 # before the constructor
    def __init__(self,name,age):
        self.__name = name
        self.age = age
        Dog.counter += 1


dog1 = Dog('Mike',5)
dog2 = Dog('Cisco',6)
dog3 = Dog('French',7)

print(dog1.counter)
print(dog2.counter)
print(dog3.counter)
```

## hasattr() property

```python
class Dog:
    counter = 0 # before the constructor
    def __init__(self,name,age):
        self.__name = name
        self.age = age
        Dog.counter += 1

bom = Dog('BombadilTom',999)
```

```python
if hasattr(bom,'name'):
    print('His name is',bom.name)
else:
    print('No name')
```
