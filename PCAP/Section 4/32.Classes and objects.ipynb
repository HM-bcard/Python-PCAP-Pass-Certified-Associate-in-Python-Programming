### Classes can contain two types of entities: variables called 'attributes', and functions calles 'methods'

# A3 (Write an) example of a class:

```python
class User: #camelcase
    def __init__(self,): # constructor
        self.nickname = 'nickname'
        self.city = 'warsaw'
    def introduce(self):
        print("Hi, i'm ",self.nickname,"from ",self.city )
        
```

# A4

## Creating an object from a class is called 'instantiating'

### __init__ keyword is required for a class to have a constructor, and the method with the \_\_init\_\_ method is treated as the constructor
### constructor needs at least one parameter - traditionally it's called 'self'

# A5
## Instatiation example

```python
sample_User = User()
```

# A6

## Calling a method in the sample instance

```python
sample_User.introduce()
```

# A7
## Calling inner entities of an object

```python
print(sample_User.nickname)
```
