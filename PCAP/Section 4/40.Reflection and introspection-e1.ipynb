## Example: a function that turns any string into an empty string

```python
def empty_strings(iobject): # an object input
    for var in iobject.__dict__.keys(): # objects dictionary keys - a standard dict method
        var_val = getattr(iobject,var) # takes object and property
        if isinstance(var_val,str): # if it's of a given type
           setattr(iobject,var,'') # takes 3 values: the object passed, the attribute name, the new attributes value
            
```
