## syntaxof simple multiple inheritance:

```python
class Stationery():
    def write(self):
        print("blablabla")

    def what_am_i(self):
        print("i am a ",Stationery.__name__)

class Cylinder():
    def shape(self):
        print("i am cylindrical")

    def what_am_i(self):
        print("i am a ",Cylinder.__name__)

class pen(Stationery,Cylinder):
    pass
```

```python
parker = pen()
parker.write()
parker.shape()
```

```python
parker.what_am_i()
```

## Caveat - each language has a different one

# Python's MRO:

### 1.Object
### 2.Superclass from left to right
### 3. Error

```python
class Stationery():
    def write(self):
        print("blablabla")

    def what_am_i(self):
        print("i am a ",Stationery.__name__)

class Cylinder():
    def shape(self):
        print("i am cylindrical")

    def what_am_i(self):
        print("i am a ",Cylinder.__name__)

class pen(Cylinder,Stationery):
    pass
```

```python
parker = pen()
parker.what_am_i()
```

### left to right

```python
class Stationery():
    def write(self):
        print("blablabla")

    def what_am_i(self):
        print("i am a ",Stationery.__name__)

class Cylinder():
    def shape(self):
        print("i am cylindrical")

    def what_am_i(self):
        print("i am a ",Cylinder.__name__)

class Pen(Cylinder,Stationery): # object->right->left
    def what_am_i(self):
        print("i am a ",Pen.__name__)
```

```python
parker = pen()
parker.what_am_i() 
```
