## Example of a class and an instance

```python
class WareHouse:
    counter = 0
    def __init__(self,address,space,cost):
        self.address = address
        self.space = space
        self.cost = cost
        WareHouse.counter += 1

    def display(self):
        print('The warehouse is located at: ',self.address,'with a space of: ',self.space,"it's cost is ", self.cost)

jersey_w1 = WareHouse('Bloomingfield 12 av.',2000,500_000)
jersey_w1.display()
print(WareHouse.counter)
```

```python
class WareHouse:
    counter = 0
    def __init__(self,address,space,cost):
        self.address = address
        self.space = space
        self.cost = cost
        WareHouse.counter += 1

        WareHouse.quality = 'low' # class variable belongs to the class not the object
        self.quality = "medium" # object variable
        quality = "high"

        # accessing from the constructor
        print(WareHouse.quality, self.quality,quality)

    def display(self):
        print('The warehouse is located at: ',self.address,'with a space of: ',self.space,"it's cost is ", self.cost)

jersey_w1 = WareHouse('Bloomingfield 12 av.',2000,500_000)
jersey_w1.display()
print(WareHouse.counter)
```
