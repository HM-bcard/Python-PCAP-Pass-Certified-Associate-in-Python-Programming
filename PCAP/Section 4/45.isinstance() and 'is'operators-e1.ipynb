## isinstance() and is()
## What does isinstance() do?

### It's used to check thether an object has been instantiated using a chosen class

```python
class Animal:
    def __init__(self,speed):
        self.speed = speed
class LandAnimal(Animal):
     def __init__(self,speed,paws):
         super().__init__(speed)
         self.paws = paws
class Dog(LandAnimal):
    pass
```

```python
twitty = Animal(20)
Tom = LandAnimal(15,4)
Fluffy = Dog(10,4)
```

```python
print(isinstance(twitty,Dog))
print(isinstance(Tom,Dog))
print(isinstance(Fluffy,Dog))
```
