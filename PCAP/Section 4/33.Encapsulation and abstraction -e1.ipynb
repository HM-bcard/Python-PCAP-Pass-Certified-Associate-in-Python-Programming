## adding methods impacting the attributes:

```python
class Car:
    def __init__(self,model,color,initial_speed = 60):
        self.model = model
        self.color = color
        self.speed = initial_speed
        
    def speed_up(self):
        self.speed += 1

    def slow_down(self):
        self.speed -= 1

    def speedometer(self):
        print('kph: ', self.speed)

        
```

```python
my_car = Car('accord','black',100)
my_car.speed_up()
my_car.speed_up()
my_car.speedometer()
```

# A4 

## Creating method constraints for realistic real world scenarios

```python
class Car:
    def __init__(self,model,color,initial_speed = 60):
        self.model = model
        self.color = color
        self.speed = initial_speed
        
    def speed_up(self):
        self.speed += 1

    def slow_down(self):
        if self.speed < 1:
            self.speed = 0
        else:
            self.speed -= 1
    def speedometer(self):
        print('kph: ', self.speed)
```

```python
my_car = Car('accord','black',0)
```

```python
my_car.slow_down()
```

```python
my_car.speedometer()
```

## modifying the constructor so it doesn't allow unrealistic values:

```python
class Car:
    def __init__(self,model,color,initial_speed = 60):
        self.model = model
        self.color = color
        if initial_speed <0:
            self.speed = 0
        else:
            self.speed = initial_speed
        
        
    def speed_up(self):
        self.speed += 1

    def slow_down(self):
        if self.speed < 1:
            self.speed = 0
        else:
            self.speed -= 1
    def speedometer(self):
        print('kph: ', self.speed)
```
