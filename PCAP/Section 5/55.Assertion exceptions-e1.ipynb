## Asserting a value, syntax:

```python
def inverse(number):
    assert (number != 0), 'Got a Zeerooo'# we assert that a value is (or isn't equal to or in range )
    return 1/number
```

```python
inverse(1)
```

```python
inverse(0)
```

```python
def inverse(number):
    assert (number in range (6)), 'Out of range'
    return 1/number
```

```python
inverse(7)
```

```python
inverse(4)
```
