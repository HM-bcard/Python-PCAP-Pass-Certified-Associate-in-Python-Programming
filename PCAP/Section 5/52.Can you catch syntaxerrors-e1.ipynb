## Can you catch  a SyntaxError in a try/except block?

### No , the only way is to manually raise it:

```python
try:
  raise SyntaxError
except:
  print('Got it!') # SyntaxError is caught here
```

```python
try:
  5:4 # this line generates a SyntaxError
except:
  print('Got it!') # SyntaxError is NOT caught here
```
