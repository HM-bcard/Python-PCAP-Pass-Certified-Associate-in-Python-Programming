## You can use else statements with try/except blocks:

```python
try:
    a = int(input("Not a zero please:"))
    print(1/a)
except ZeroDivisionError:
    print("EH, must you not listen?")
except ValueError:
    print("Not a letter too :P")
else:
    print("Very good!")
```

```python
try:
    a = int(input("Not a zero please:"))
    print(1/a)
except ZeroDivisionError:
    print("EH, must you not listen?")
else:
    print("Very good!")
```
