## How to use the raise keyword?

### It's used to manually raise an exception

```python
def bigger(a,b):
    if not isinstance(a,int) or not isinstance(b, int):
        raise ValueError # the TypeError has been overwritten by ValueError
    if b > a:
        return b
        
```

```python
def bigger2(a,b):
    if b > a:
        return b
    if a > b:
        return a
```

```python
bigger(1,a)
```

```python
bigger(1,'a')
```

## Second usage of the raise keyword is:

### Prompting python to display the full error message stack, even though we want it to print the except clause contents on the first line

```python
def reverse(a):
    try:
        return(1/a)
    except:
        print("There was an error, check the error stack in the logs below")
        raise
    
```

```python
reverse(0)
```
