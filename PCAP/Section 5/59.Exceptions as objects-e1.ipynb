# A1

## Are exceptions objects?

### yes and we can assign them and use them abstractly

```python
def bigger2(a,b):
    try:
        if b > a:
            return b
        if a > b:
            return a
    except Exception as e:
        print(e)
        raise
        return None
```

```python
bigger2('a',2)
```

## exception arguments:

```python
def bigger2(a,b):
    try:
        if b > a:
            return b
        if a > b:
            return a
    except Exception as e:
        print(e)
        print(e.args)
        raise
        return None
```

```python
bigger2(1,'z')
```
