## Differences between find() and index()

find() is safer to use, because if a value isn't found it returns '-1' not an error, 
but find works only on strings and index works also on other data types

```python
list1 = [1,2,3,4,5,6,7]
string2 = '1234567'

list1.index(1)

#list1.find(1)
```

```python
list1.find(1)
```

## rfind - right find - lists the occurence of a given substring first from the end of the string
